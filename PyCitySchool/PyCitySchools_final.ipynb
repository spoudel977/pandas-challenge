```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load (Remember to Change These)
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

```

## District Summary

```python
# Calculate the total number of unique schools

school_count = school_data_complete["school_name"].nunique()
school_count

# school_count = school_data_complete["school_name"].value_counts()
```

```python
# Calculate the total number of students
student_count = school_data_complete['student_name'].count()
student_count
```

```python
# Calculate the total budget
total_budget = school_data_complete['budget'].sum()
total_budget
```

```python
# Calculate the average (mean) math score
average_math_score = school_data_complete['math_score'].mean()
average_math_score
```

```python
# Calculate the average (mean) reading score
average_reading_score = school_data_complete['reading_score'].mean()
average_reading_score
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
passing_math_count = school_data_complete[(school_data_complete["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / float(student_count) * 100
passing_math_percentage
```

```python
# Calculate the percentage of students who passed reading (hint: look at how the math percentage was calculated)  
passing_reading_count = school_data_complete[(school_data_complete["reading_score"] >= 70)].count()["student_name"]
passing_reading_percentage = passing_reading_count / float(student_count) * 100
passing_reading_percentage
```

```python
# Use the following to calculate the percentage of students that passed math and reading
passing_math_reading_count = school_data_complete[
    (school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(student_count) * 100
overall_passing_rate
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
district_summary = pd.DataFrame({"Total School":  [school_count],
                                                   "Total Students":  [student_count],
                                                   "Total Budget": [total_budget],
                                                   "Average Math Score": [average_math_score],
                                                   "Average Reading Score": [average_reading_score],
                                                   "% Passing Math": [passing_math_percentage],
                                                   "% Passing Reading": [passing_reading_percentage],
                                                   "% Overall Passing": [overall_passing_rate]})

# Formatting
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)

# Display the DataFrame
district_summary
```

```python
# Create a DataFrame called `per_school_summary` with columns for the calculations above.
# Create an empty list to store dictionaries for each school
per_school_summary = []

# Iterate through each school in the dataset
for school_name, school_data in school_data_complete.groupby('school_name'):
    per_school_summary.append({
        'School Name': school_name,
        'School Type': school_data['type'].iloc[0],
        'Total Students': len(school_data),
        'Total School Budget': school_data['budget'].iloc[0],  # Assuming 'budget' is a column in school_data_complete
        'Per Student Budget': school_data['budget'].iloc[0] / len(school_data),
        'Average Math Score': school_data['math_score'].mean(),  # Assuming 'math_score' is a column in school_data_complete
        'Average Reading Score': school_data['reading_score'].mean(),  # Assuming 'reading_score' is a column in school_data_complete
        '% Passing Math': (school_data['math_score'] >= 70).mean() * 100,
        '% Passing Reading': (school_data['reading_score'] >= 70).mean() * 100,
        '% Overall Passing Rate': (((school_data['math_score'] >= 70) & (school_data['reading_score'] >= 70)).mean()) * 100
    })

# Convert the list of dictionaries to a DataFrame
per_school_summary_df = pd.DataFrame(per_school_summary)

# Formatting
per_school_summary_df["Total School Budget"] = per_school_summary_df["Total School Budget"].map("${:,.2f}".format)
per_school_summary_df["Per Student Budget"] = per_school_summary_df["Per Student Budget"].map("${:,.2f}".format)
per_school_summary_df.index = per_school_summary_df.index + 1


# Display the DataFrame
display(per_school_summary_df)
```

```python
# First Convert the list to a DataFrame
school_spending_df = pd.DataFrame(school_spending_df, columns=["Per Student Budget"])


# Check if "Per Student Budget" column contains numeric values
if school_spending_df["Per Student Budget"].dtype != 'float64':
    # Attempt to convert the column to numeric, coercing errors to NaN
    school_spending_df["Per Student Budget"] = pd.to_numeric(school_spending_df["Per Student Budget"], errors='coerce')

# Define the spending bins and labels
spending_bins = [0, 585, 630, 645, 675]
spending_labels = ["<$585", "$585-630", "$630-645", "$645-675"]

# Use pd.cut to categorize spending based on the bins
school_spending_df["Spending Ranges(Per Student)"] = pd.cut(school_spending_df["Per Student Budget"], bins=spending_bins, labels=spending_labels)

# Assuming per_school_summary_df contains the column "School Name"
# Assign the school names to the "School Name" column
school_spending_df["School Name"] = per_school_summary_df["School Name"]

# Merge per_school_summary_df with school_spending_df on the "School Name" column
school_spending_df = pd.merge(school_spending_df, per_school_summary_df[["School Name", "School Type", "Total Students", "Total School Budget", "Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "% Overall Passing Rate"]], on="School Name", how="left")

# Move the "School Name" column to the second position
school_spending_df = school_spending_df[['School Name'] + [col for col in school_spending_df.columns if col != 'School Name']]

# Fill missing values for row with index 0
school_spending_df.fillna(method='bfill', inplace=True)

# Display the DataFrame
school_spending_df


```

## Scores by School Size

```python
# Establish the bins.
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
# Define School Size
school_size = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

# Create DataFrame with unique school names
school_size_summary_df = per_school_summary_df.copy()  # Create a copy of per_school_summary_df

# Check for any non-numeric characters in 'Total School Budget' and 'Total Students' columns
non_numeric_budget = school_size_summary_df['Total School Budget'].str.replace(',', '').str.isnumeric().all()
non_numeric_students = school_size_summary_df['Total Students'].astype(str).str.isnumeric().all()

# If there are any non-numeric values, print a warning message and attempt to convert them
if not non_numeric_budget or not non_numeric_students:
    print("Warning: Non-numeric values found in 'Total School Budget' or 'Total Students' columns. Attempting to convert them...")

    # Attempt to remove non-numeric characters and convert to numeric
    school_size_summary_df['Total School Budget'] = pd.to_numeric(school_size_summary_df['Total School Budget'].str.replace('[^\d.]', '', regex=True), errors='coerce')
    school_size_summary_df['Total Students'] = pd.to_numeric(school_size_summary_df['Total Students'].astype(str).str.replace('[^\d.]', '', regex=True), errors='coerce')

    # Check if conversion was successful
    if school_size_summary_df[['Total School Budget', 'Total Students']].isnull().any().any():
        print("Warning: Unable to convert some values to numeric. Please check the data.")

# Calculate per student budget for each school
school_size_summary_df["Per Student Budget"] = school_size_summary_df['Total School Budget'] / school_size_summary_df['Total Students']

# Use pd.cut to categorize size based on the bins of Total Students
school_size_summary_df["School Size"] = pd.cut(school_size_summary_df["Total Students"], bins=size_bins, labels=school_size)

# Drop duplicate rows based on "School Name"
school_size_summary_df.drop_duplicates(subset="School Name", keep="first", inplace=True)

# Display the DataFrame
school_size_summary_df
```
